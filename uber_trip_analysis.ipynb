```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("uber_trip.csv")
```

```python
df.head()
```

```python
total_trips = df["Trip_ID"].count()
print(total_trips)
```

```python
total_revenue = df["Total_Fare"].sum()
print(total_revenue)
```

```python
avg_fare = df["Total_Fare"].mean()
print(avg_fare)
```

```python
avg_rating = df["Customer_Rating"].mean()
print(avg_rating)
```

```python
total_distance = df["Distance_km"].sum()
print(total_distance)
```

```python
# Revenue by city
df.groupby("Pickup_City")["Total_Fare"].sum()
```

```python
# trips by city
df.groupby("Pickup_City")["Trip_ID"].count()
```

```python
df["Month"] = pd.to_datetime(df["Date"]).dt.month_name()
df.groupby("Month")["Trip_ID"].count()
```
